```python
from langchain.agents import create_agent
from langchain.tools import tool, ToolRuntime #### <--- import this for langchain tools
```

```python
import os
os.environ["GOOGLE_API_KEY"] = "----insert your google api key here----"
```

```python
from tavily import TavilyClient
```

```python
@tool
def tavilySearchEng(query: str):
  """web search"""
  client = TavilyClient(api_key = "----insert your tavily api key here----")

  response = client.search(
      query = query,
      max_results = 5
  )

  print(response)
```

```python
agent1 = create_agent(
    model="google_genai:gemini-2.5-flash-lite",
    tools=[tavilySearchEng],
    system_prompt="You are a helpful assistant",
)
```

```python
result = agent1.invoke(
    {"messages": [{"role": "user", "content": "what is news in india today"}]}
)
print(result["messages"][-1].content_blocks)
```
