```python
from langchain.agents import create_agent
from langchain.tools import tool
```

```python
import os
os.environ["GOOGLE_API_KEY"] = "--- google api key here ---"
```

```python
@tool
def getmethemenu():
  ''' this function is called whenever menu is being asked '''
  return '''Today's menu includes:

- Masala Dosa
- Idli Sambar
- Medu Vada
- Pongal
- Filter Coffee'''
```

```python
system_prompt1 = '''
you are a professional chef, with 10+ years of exeprience in the field. serving all the country India, You have a great expertise in south indian cuisine, you are
one of top 10 chefs in india, whenever a question is asked to you, answer precisely dont over explain answer to the point.

When tools provide information,
ALWAYS use the tool output directly in your final answer.

Never say information is unavailable if a tool has already returned data.
'''
```

```python
model = create_agent(
    model = "google_genai:gemini-2.5-flash-lite",
    tools = [getmethemenu],
    system_prompt = system_prompt1
)
```

```python
invoke_param = {
    "messages":[{
        "role":"user",
        "content":"what is the menu?"
    }]
}
result = model.invoke(invoke_param)
```
